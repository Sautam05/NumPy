## Problem Statement
Perform concatenate and vstack on two arrays.

## Sample Input
```python
Use NumPy arrays
```

## Sample Output
```python
Expected Output
```


````python
# Write your code here
import numpy as np
a = np.array([2,4,6])
b = np.array([8,10,12])
print(np.concatenate([a,b]))
````

````python
A = np.array([[2,4,6],[8,10,12]])
B = np.array([[1,3,5],[7,9,11]])
print(np.vstack([A,B]))
````
